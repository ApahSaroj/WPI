```python
import h2o
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/DATA/WQC_test.xlsx')
df
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df
```

```python
df.drop(df.columns[df.columns.str.contains('WQC',case = False)],axis = 1, inplace = True)
df
```

```python
data = df
data['target'] = df['WPI']

h2o.init()

h2o_train = h2o.H2OFrame(data)
h2o_train
```

```python
data.drop(data.columns[data.columns.str.contains('WPI',case = False)],axis = 1, inplace = True)
data
```

```python
import h2o
from h2o.automl import H2OAutoML


regression_data = data
regression_data['target'] = data['target']

regression_data
```

```python
h2o.init()

h2o_regression_train = h2o.H2OFrame(regression_data)
h2o_regression_train['target'] = h2o_regression_train['target'].asnumeric()

h2o_regression_aml = H2OAutoML(max_runtime_secs=60, nfolds=0)
h2o_regression_aml.train(y='target', training_frame=h2o_regression_train)
```

```python
h2o_regression_aml.leaderboard.head(rows=h2o_regression_aml.leaderboard.nrows)
```

```python
h2o_regression_aml.leader
```

**Lets use multi-level classification means considering WQC**

```python
df1 = pd.read_excel('/content/drive/MyDrive/DATA/WQC_test.xlsx')
df1
```

```python
df1.drop(df1.columns[df1.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df1
```

```python
import h2o
from h2o.automl import H2OAutoML


multiclass_data = df1
multiclass_data['target'] = df1['WQC']

multiclass_data
```

```python
multiclass_data.drop(multiclass_data.columns[multiclass_data.columns.str.contains('WQC',case = False)],axis = 1, inplace = True)
multiclass_data.drop(multiclass_data.columns[multiclass_data.columns.str.contains('WPI',case = False)],axis = 1, inplace = True)
multiclass_data
```

```python
h2o.init()

h2o_multiclass_train = h2o.H2OFrame(multiclass_data)
h2o_multiclass_train['target'] = h2o_multiclass_train['target'].asfactor()

h2o_multiclass_aml = H2OAutoML(max_runtime_secs=60, nfolds=0)
h2o_multiclass_aml.train(y='target', training_frame=h2o_multiclass_train)
```

```python
h2o_multiclass_aml.leaderboard.head(rows=h2o_multiclass_aml.leaderboard.nrows)
```

```python
h2o_multiclass_aml.leader
```
